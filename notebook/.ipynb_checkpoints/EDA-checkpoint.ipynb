```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
plt.style.use('ggplot')
```

```python

# Load Dataset

file_path = "../data/imdb-movies-dataset.csv" # chỉnh lại nếu cần
df = pd.read_csv(file_path)
# Chuẩn hóa tên cột
df.columns = [c.lower().replace(' ', '_') for c in df.columns]
df.head()
```

```python
# Giữ lại các cột quan trọng
cols_to_keep = ['title', 'year', 'genre', 'rating', 'description', 'director', 'cast', 'poster','review']
df_slim = df[cols_to_keep].copy()

print(df_slim.info())
```

```python
#Xử lý Trùng lặp (Duplicates)
print(f"Số dòng trước khi xử lý trùng lặp: {len(df_slim)}")
df_slim.drop_duplicates(subset=['title', 'year'], inplace=True)
print(f"Số dòng sau khi xử lý trùng lặp: {len(df_slim)}")
```

```python
#Kiểm tra Missing Values
df_slim.isnull().sum()
```

```python
# xử lý các giá trị thiếu theo cột 
# cột description Bắt buộc phải không thiếu và xóa giá trị còn thiếu của cột cast và director  
df_slim.dropna(subset=['description','cast','director'], inplace=True)


#review Điền giá trị rỗng
df_slim['review'] = df_slim['review'].fillna("")

# cột genre điền giá trị Unknown
df_slim['genre'] = df_slim['genre'].fillna("Unknown")

#cột rating và year Điền Trung vị
medians = df_slim[['rating', 'year']].median()
# Cột rating: Điền Trung vị rating
df_slim['rating'] = df_slim['rating'].fillna(medians['rating'])
# Cột year: Điền Trung vị year và chuyển sang kiểu số nguyên (int)
df_slim['year'] = df_slim['year'].fillna(medians['year']).astype(int)

print("\nKiểm tra cuối cùng (Tỷ lệ thiếu giá trị):")
print(df_slim.isnull().sum())
```

```python

#  Kiểm tra thông tin tổng quát
df_slim.info()
df_slim.describe(include='all')
```

```python
# phân tích độ dài từ 
df_slim['word_count'] = df_slim['description'].apply(lambda x: len(str(x).split()))

print("Thống kê Độ dài Từ:")
print(df_slim['word_count'].describe())

# Vẽ histogram
plt.figure(figsize=(10, 5))
sns.histplot(df_slim['word_count'], bins=30, kde=True, color='blue')
plt.title('Phân phối Độ dài Từ trong Mô tả phim')
plt.show()
```

***Độ dài trung bình ~26 từ cho thấy mô tả phim rất ngắn.  Dữ liệu ngắn  kết hợp thêm title và genre vào văn bản để làm giàu ngữ cảnh trước khi tạo vector SBERT/TF-IDF***

```python
df_genre = df_slim.dropna(subset=['genre']).copy()
genre_series = df_genre['genre'].str.split(', ').explode()

plt.figure(figsize=(10, 6))
genre_series.value_counts().head(10).plot(kind='barh', color='darkgreen')
plt.title('Top 10 Thể loại Phổ biến nhất')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Rating
sns.histplot(df_slim['rating'].dropna(), bins=20, kde=True, ax=axes[0], color='purple')
axes[0].set_title('Phân phối Điểm Rating')

# Year
sns.histplot(df_slim['year'].dropna(), bins=30, kde=False, ax=axes[1], color='orange')
axes[1].set_title('Phân phối Năm phát hành')

plt.tight_layout()
plt.show()
```

```python
# Bỏ cột 'word_count' đã tạo cho EDA
df_final = df_slim.drop(columns=['word_count'])

# Lưu vào thư mục data
df_final.to_csv('../data/clean_movies.csv', index=False)
print("Đã hoàn thành EDA và lưu file clean_movies.csv.")
```
